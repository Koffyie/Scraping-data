**Vectorization.**

We've been interacting with data in its textual form but machines don't understand that. Computers understand 0s and 1s, numbers are the most direct way to communicate with them. The process of converting text into numeric data is called Vectorization.

Introducing: Bag of Words

This a method of representing text data where the order or structure of words is discarded, and only the frequency or presence of words is considered.

One-hot Encoding

A one-hot encoding is a vector that assigns each word to an index such that when it is being processed, all other indexes representing the other words in the dataset are 0s while the word of interest is a 1. For instance, if the only sentence in the dataset was: "The ultimate answer is 42." then, the vector corresponding to answer might have the vector [0, 0, 1, 0, 0]. In this case, indices match the occurence of the word in the sentence).

```python
!pip install scikit-learn
```

```python
from sklearn.preprocessing import OneHotEncoder

#Create a vocabulary from the tokenized dataset
vocab = sorted(set(token for sentence in wt_quotes_lines for token in sentence))

#Use OneHotEncoder to One-hot encode the vocabulary
encoder = OneHotEncoder(categories=[vocab], sparse_output=False)

#One-hot encode the tokenized sentences
one_hot_encoded_quotes = []

for sentence in tqdm(wt_quotes_lines):
  encoded_sentence = encoder.fit_transform([[token] for token in sentence])
  one_hot_encoded_quotes.append(encoded_sentence)
```

TF-IDF

One-hot encodings are nice but they don't hold a lot of contextual information about what a word is in a sentence and how it is reflected in other sentences or across the whole dataset. One way to escape this limitation is by creating term frequency-inverse document frequency vectors instead

Term Frequency

This measures how frequently a term occurs in a document. In our context, documents are sentences. If we have a document with 100 words and the term "apple" appears 5 times, the term frequency for "apple" is 5/100 = 0.05.

```python
def compute_tf(document, word):
  return document.count(word) / len(document)

```

```python
wt_quotes_lines[10]
```

```python
compute_tf(wt_quotes_lines[10], "life")
```

```python
compute_tf(wt_quotes_lines[10], "you")
```

Inverse Document Frequency

This measures how unique or rare a term is across all documents in the corpus. If the term "apple" appears in 50 out of 1000 documents, the IDF is calculated as the logarithm of the total number of documents divided by the number of documents containing the term.

```python
def compute_idf(word, quotes_lines):
    docs_containing_word = sum(1 for sentence in quotes_lines if word in sentence)
    return np.log(len(quotes_lines) / (1 + docs_containing_word))  # Added 1 to avoid division by zero
```

```python
compute_idf("you", wt_quotes_lines)
```

```python
compute_idf("love", wt_quotes_lines)
```

Bringing TF and IDF together

The TF-IDF is simply the product of TF and IDF. It's designed to reflect how important a word is to a document in a dataset (or corpus).



```python
def compute_tf_idf(document, word, wt_quotes_lines):
    tf = compute_tf(document, word)
    idf = compute_idf(word, ct_quotes_lines)
    return tf * idf
```

```python
def sentence_tf_idf(wt_sentence, vocab):
    # Calculate TF-IDF for a sentence
    vector = []
    for word in wt_sentence:
        vector.append(compute_tf_idf(wt_sentence, word, wt_quotes_lines))
    return vector
```

```python
# Display results
idx = 10
wt_sentence = wt_quotes_lines[idx]
print("Sentence:", ' '.join(wt_sentence))

vector = sentence_tf_idf(wt_sentence, vocab)
for i, score in enumerate(vector):
    print(f"{wt_sentence[i]}: {score:.4f}")
```

Gensim TF-IDF

We don't have to do all this ourselves, there's a library called Gensim that can do all this for us, allowing us to interact with language models.

```python
!pip install gensim
```

```python
import gensim as gs
from gensim import corpora
from gensim.models import TfidfModel
```

```python
# Create a dictionary from the tokenized sentences
dictionary = corpora.Dictionary(wt_quotes_lines)

# Convert tokenized sentences into bag-of-words format
bow_corpus = [dictionary.doc2bow(sentence) for sentence in wt_quotes_lines]

# Create a TF-IDF model
tfidf_model = TfidfModel(bow_corpus)

# Get TF-IDF vectors
tfidf_vectors = [tfidf_model[bow] for bow in bow_corpus]
```

```python
# Display results
idx = 10
sentence = wt_quotes_lines[idx]
print("Sentence:", sentence)
for word_id, score in tfidf_vectors[idx]:
    print(f"{dictionary[word_id]}: {score:.4f}")
```

Comparing Sentences

To compare sentences, we would typically have to look at them and spot differences in interpretation based on our understanding of the context. Since we now have the sentences as numerical vectors, one way to compare them is using cosine similarity.

cosine similarity(A,B)=A⋅B|∥A∥∥B∥

```python
sample_idx_1 = 5
sample_idx_2 = 10

print(f"Sentence 1: {' '.join(wt_quotes_lines[sample_idx_1])}")
print(f"Sentence 2: {' '.join(wt_quotes_lines[sample_idx_2])}")

similarity = gs.matutils.cossim(
    vec1=tfidf_vectors[sample_idx_1],
    vec2=tfidf_vectors[sample_idx_2],
)
print(similarity)
```

```python
from gensim.models import KeyedVectors
word_vectors = KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin', binary=True)
```

```python
word_vectors.key_to_index
```

```python
def get_gensim_embeddings(wt_sentence):
    embeddings = []
    for word in wt_sentence:
        if word in word_vectors:
            embeddings.append(word_vectors[word])
    return embeddings
```

```python
gensim_embeddings = get_gensim_embeddings(wt_quotes_lines[0])
print(np.shape(gensim_embeddings))
```

Word Similarityy

```python
def gensim_similarity(word1, word2):
    if word1 in word_vectors and word2 in word_vectors:
        return word_vectors.similarity(word1, word2)
    return 0
```

```python
print(gensim_similarity('life', 'love'))
```

```python
print(gensim_similarity('religion', 'God'))
```

```python
selected_words = ["love", "hate", "friendship", "artificial", "language", "word", "science", "book", "paper", "constitution", "law"]
word_embeddings = np.array([word_vectors[word] for word in selected_words if word in word_vectors])
word_embeddings.shape            #Visualizing embeddings
```

```python
from sklearn.manifold import TSNE

def visualize_embeddings_2D(selected_words, word_embeddings, word_vectors):
    # Project embeddings into 2D using t-SNE
    tsne = TSNE(n_components=2, random_state=42, learning_rate='auto', perplexity=10)
    word_2d_embeddings = tsne.fit_transform(word_embeddings)

    plt.figure(figsize=(8, 8))
    for i, word in enumerate(selected_words):
        if word in word_vectors:
            plt.scatter(word_2d_embeddings[i, 0], word_2d_embeddings[i, 1], marker='+', color='red')
            plt.text(word_2d_embeddings[i, 0]+0.2, word_2d_embeddings[i, 1], word, fontsize=9)
    plt.show()
```

```python
visualize_embeddings_2D(selected_words, word_embeddings, word_vectors)
```
