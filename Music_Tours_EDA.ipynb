```python
#Creating a wrangling function
def wrangle_data(data_path):
  dt = pd.read_csv(data_path)
  dt = dt.drop(columns=["Rank"])
  chars_to_remove = [',', '$', '[b]', '[e]', '[]']
  for char in chars_to_remove:
    dt['Actual\xa0gross'] = dt['Actual\xa0gross'].apply(lambda x: x.replace(char, ''))
    dt['Average gross'] = dt['Average gross'].apply(lambda x: x.replace(char, ''))
  dt['Adjusted\xa0gross (in 2022 dollars)'] = dt['Adjusted\xa0gross (in 2022 dollars)'].str.replace('[,$]', '', regex=True)ssss
  dt['Actual\xa0gross'] = dt['Actual\xa0gross'].astype(float)
  dt['Average gross'] = dt['Average gross'].astype(float)
  dt['Adjusted\xa0gross (in 2022 dollars)'] = dt['Adjusted\xa0gross (in 2022 dollars)'].astype(float)


  return dt
```

```python
dt = wrangle_data('/content/my_file (1).csv')
dt.head(20)
```

```python
artist_shows = dt.groupby('Artist')['Shows'].sum().sort_values(ascending=False)
print(artist_shows)
```

```python
artist_shows.values
```

```python
#pandas plotting function
plt.figure(figsize=(12, 6))
artist_shows.plot(x = artist_shows.index, y = artist_shows.values, kind = 'bar', title = 'Number of Shows by Artist')
```

```python
#using seaborn
fig, ax = plt.subplots()
fig.set_size_inches(12, 6)

ax = sns.barplot(x=artist_shows.index, y=artist_shows.values)
plt.xlabel('Artist')
plt.ylabel('Number of Shows')
plt.title('Barplot showing the Number of Shows per Artist');

```

```python
#Using plotly
fig = px.bar(
    x = artist_shows.index,
    y = artist_shows.values,
)

fig.update_layout(
    title = 'Number of Shows per Artist',
    xaxis_title = 'Artist',
    yaxis_title = 'Number of Shows',
)

fig.show()
```

```python
#What artist has the highest gross overall
artist_actual_gross = dt.groupby('Artist')['Actual\xa0gross'].sum().sort_values(ascending=False).reset_index()
print(artist_actual_gross)
```

```python
fig = px.bar(artist_actual_gross,x = "Actual\xa0gross", y = "Artist", color = "Artist", title = "Total Gross of Artists")
fig.show()
```

```python
# Extract the starting year (if range, take the first year)
dt['Year_start'] = dt['Year(s)'].astype(str).str.extract(r'(\d{4})').astype(float)

tours_per_year = dt.groupby('Year_start').size()

tours_per_year.plot(kind='line', marker='o', figsize=(8,5), title="Number of Tours per Year")
plt.xlabel("Year")
plt.ylabel("Number of Tours")
plt.show()
```
